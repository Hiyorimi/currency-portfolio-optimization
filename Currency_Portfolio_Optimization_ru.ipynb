## Введение

Оптимизация инсвестиционного портфеля — это одна из базовых проблем, с которой сталкиваются люди, желающие инвестировать деньги _(или любые другие виды капитала, такие как время)_. Наиболее частой областью инвестиций является рынок ценных бумаг,на котором управляющий портфелем старается выполнить две цели:

1. Увеличить доходность
2. Уменьшить риск 

Увеличение доходности означает выбор таких объектов для инсветиций, ожидание доходности которых будет максимальным. А уменьшение риска означает, что выбранные объекты с большей вероятностью оправдают наши ожидания, чем будут убыточными.

Все люди, которые открывали пробный счёт на форексе или хоть немного понимают, как устроена жизнь, знают, что эти цели по сути дела диметрально противоположны. Все инестиции с высоким уровнем доходности зачастую имеют большой разброс. Так что задача оптимизации портфолио инвестиций — это задача нахождения баланса между этими двумя целями.


## Игра на разнице курса валют

В этом посте мы представим себя на месте инвестиционного менеджера _(например, управляющего портфолио состоятельных клиентов)_. Мы хотим вложить деньги в иностранные валюты так, чтобы получить максимум прибыли через месяц в долларах США. Нас не интересуют абсолютные величины, мы лишь хотим максимизировать процент прибыли после покупки иностранной валюты.

Для начала мы скачаем историю курса за последние 100 месяцев с портала Федеральной Резервной системы США [FRB H10: Data Download Program](http://www.federalreserve.gov/datadownload/Build.aspx?rel=H10). Вы также можете скопировать [репозиторий, содержащий этот пост](https://github.com/yhat/currency-portfolio-optimization), в котором расположены оригинальный ipynb-файл и исходные данные.

Анализируя данные, мы заметим кое-что странное. Первые 6 записей содержат общие сведения о данных, помимо этого мы не будем отображать данные о валюте, помеченной как  `NA`.

```python
import csv
import pandas

raw_data = [row for row in csv.reader(open('2014-12-28-fed-monthly-currency-data.csv'))]

# Transpose header metadata and turn it into a data frame
head_data = [[] for _ in range(len(raw_data[0]))]
for row in raw_data[:6]:
    for j, val in enumerate(row):
        head_data[j].append(val)
    
metadata = pandas.DataFrame.from_records(data=head_data[1:], columns=head_data[0])

# Filter out NA currencies
metadata = metadata[metadata['Currency:'] != 'NA']
metadata
```

Для удобства мы сделаем отдельный список для названий стран и валют. Это позволит нам понимать, какую информацию о валютах мы получили.

```python
# This will have the same indices as the rest of the metadata.
countries = []
for c in metadata['Series Description']:
    c = c.upper()
    c = c.split(' -- ')[0]
    c = c.split(' - ')[0]
    countries.append(c)
countries[1] = 'EURO AREA'
countries
```

Также мы выведем информацию о разнице курсов валют в каждый месяц по отношению к курсу доллара.

```python
exchange_rates = pandas.DataFrame.from_records(data=raw_data[6:], columns=raw_data[5])
exchange_rates = exchange_rates[['Time Period'] + list(metadata['Time Period'])]
exchange_rates.tail()
```

## Вычисление результатов

Сейчас у нас есть набор данных, описывающих движение курса по отношению к курсу долларов США. Но мы должны рассчитать % прибыли  на каждый купленный доллар, который мы получим в конце месяца.

Чтобы расчитать возврат инвестиций, мы должны перевести каждое значение `курса валют` в количество местной валюты, за которое мы можем купить \$1. Большая часть данных уже представлена в этом виде, но четыре страны: Австралия, Еврозона, Новая Зеландия и  Великобритания записаны в долларах. Мы возьмём обратные величины, разделив `1` на эти значения, чтобы получить количество местной валюты, на которое мы можем купить 1 доллар.

$$\$1 = x \rightarrow \frac{1}{x} = \frac{1}{\$}$$ 

Теперь мы рассчитаем % прибыли на каждый месяц в долларах США. Так что если $r_m$ — курс обмена валют в месяц $m$, то $\frac{r_m - r_{m+1}}{r_{m+1}}$  — это % прибыли, который мы получим, продав валюту в конце месяца $m$. К примеру, в месяце $m$  мы можем купить 2 единицы валюты за \$1. В следующий месяц $m+1$, \$1 позволит купить 4 единицы этой валюты. Тогда доходность этой операции будет $\frac{2 - 4}{4} = -50\%$.

```python
# Convert the exchange rate data frame into percentage returns.
rows = []
for i in range(len(exchange_rates)-1):
    row = {}
    for tp, cur in zip(metadata['Time Period'], metadata['Currency:']):
        x1 = float(exchange_rates[tp][i])
        x2 = float(exchange_rates[tp][i+1])
    
        if cur == 'USD':
            x1 = 1.0 / x1
            x2 = 1.0 / x2
        
        # Returns are in units of %.
        row[tp] = 100 * (x1 - x2) / x2
    rows.append(row)

returns = pandas.DataFrame(data=rows, columns=list(metadata['Time Period']))
returns.tail()
```

Мы скачали исторические данные о разнице курсов валют за 100 месяцев для каждой страны. Теперь все строки таблицы содержат информацию о % прибыли. Мат. ожидания и дисперсии эти значений показывают ожидаемый возврат инвестиций в эти валюты.

Заметим, что ожидаемый возврат инвестиций (средние значения) описываются % возврата. Это означает, что 0.15 = 0.15% прибыли через месяц, что равно 1.8% за год, что вполне неплохо для инвестиций в валюту.

```python
# Means are the expected returns for each currency.
exp_returns = pandas.concat({'mean': returns.mean(), 'variance': returns.var()}, axis=1)
exp_returns
```

## Собираем инвестиционный портфель

Допустим, мы хотим увеличить возврат инвестиций в долларах США. Самым простым способом будет выбрать наиболее выгодную валюту и купить её на всю сумму инвестиций.

```python
countries[list(exp_returns.index).index(exp_returns['mean'].idxmax())]
```

Неудивительно, что швейцарский франк, являющийся наиболее стабильной валютой, может приносить 0.2% прибыли ежемесячно. Данные говорят о том, что в результате мы получим 95% доверительный интервал и ожидаемая прибыль будет между -0.2% до 0.8%.

```python
import math
n = len(exchange_rates)
sz_mean = exp_returns['mean']['RXI_N.M.SZ']
sz_sd = math.sqrt(exp_returns['variance']['RXI_N.M.SZ'])
sz_ci = (sz_mean-1.96*sz_sd/math.sqrt(n), sz_mean+1.96*sz_sd/math.sqrt(n))
print 'Monthly expected return (Swiss Francs): %0.3f%%' % sz_mean
print '95%% confidence interval on expected return: (%0.3f%%, %.03f%%)' % sz_ci
```

Важно отметить, что мы рассчитываем _ожидаемый возврат инвестиций_, но не гарантированный возврат инвестиций. Это означает, что реальная прибыль может быть отрицательной или положительной. Так что решение вложить все деньги в швейцарский франк может привести только к потери 0.2% инвестируемой суммы каждый месяц.

При этом, естетсвенно, мы можем добиться лучших результатов, для чего мы воспользуемся информацией о том, насколько курсы валют меняются относительно друг друга с течением времени. Эта идея лежит в основе [портфельной теории](http://en.wikipedia.org/wiki/Modern_portfolio_theory), которую описал Гарри Марковиц в 50х годах. Мы можем использовать ковариционную матрицу возврата инвестиций для того, чтобы хэджировать инвестиции относительно друг друга. Так что теперь перед нами стоит задача выбора инвестиций таким образом, чтобы максимизировать прибыль, соблюдая выбранный уровень риска.

```python
# Get the covariance matrix for our returns. This is normalized by N-1.
# As expected, the diagonal contains the variance for each individual currency.

returns_cov = returns.cov()
returns_cov.head()
```

Воспользуемся моделью Марковица. Согласно этой модели, $\mu$ – вектор ожидаемой прибыли для каждой валюты. $x$ — вектор неотрицательных значений с суммой равной 1, который описывает долю инвестиций в каждую валюту. Первый член функции показывает ожидаемый возврат от этой инвестиции. Без использования этой модели мы могли бы просто вложить все деньги в швейцарский франк.

$$\max{\mu^T x - \alpha x \Sigma x}$$
$$e^T x = 1$$
$$x \ge 0$$

Второй член описывает хэджирование. $\alpha$ — мера аверсии риска, которая означает следующие: когда $\alpha$ равна 0, мы инвестируем только в швейцарские франки. По мере роста $\alpha$, мы хэджируем большую часть нашего портфеля. $\Sigma$ — ковариационная матрица, тогда $x \Sigma x$ будет [вариативностью портфолио](http://en.wikipedia.org/wiki/Variance#Weighted_sum_of_variables). Таким образом, мы сможем найти оптимальное соотношение между ожидаемой прибылью и вариативностью данных.

А задача оптимизации, это — известная задача [квадратичного программирования](http://en.wikipedia.org/wiki/Quadratic_programming), так что мы можем использовать [qp](http://cvxopt.org/userguide/coneprog.html#quadratic-programming) функцию пакета [CVXOPT](http://cvxopt.org) для того, чтобы решить её. Для этого мы используем [ScienceOps](http://docs.yhathq.com/scienceops/intro/index.html) — продукт компании, в которой и написали этот пост, который представляет из себя облачную песочницу для алгоритмов дата майнинга.

```python
from cvxopt import matrix, solvers
from yhat import Yhat, YhatModel, preprocess
import numpy

class CurrencyPortfolio(YhatModel):
    REQUIREMENTS = ["cvxopt","numpy"]
    
    @preprocess(in_type=dict, out_type=dict)
    def execute(self, data):
        P = matrix(data['risk_aversion'] * returns_cov.as_matrix())
        q = matrix(-exp_returns['mean'].as_matrix())
        G = matrix(0.0, (len(q),len(q)))
        G[::len(q)+1] = -1.0
        h = matrix(0.0, (len(q),1))
        A = matrix(1.0, (1,len(q)))
        b = matrix(1.0)

        solution = solvers.qp(P, q, G, h, A, b)
        expected_return = exp_returns['mean'].dot(solution['x'])[0]
        variance = sum(solution['x'] * returns_cov.as_matrix().dot(solution['x']))[0]

        investments = {}
        for i, amount in enumerate(solution['x']):
            # Ignore values that appear to have converged to 0.
            if amount > 10e-5:
                investments[countries[i]] = amount*100

        return {
            'risk_aversion': data['risk_aversion'],
            'investments': investments,
            'expected_return': expected_return,
            'variance': variance
        }
```

Модель вернёт нам оцененный процент прибыли, названия стран и ковариационную матрицу для них. Всё что нужно, чтобы сделать это передать данные $\alpha$, которые в коде называются `risk_aversion`. Вы можете сами склопировать репозиторий и зарегистрироваться на сайте (регистрация, к слову очень вменяемая и занимает около 3х минут, причём сразу по входу в аккаунт ты видишь требуемый апи-ключ. Исходный репозиторий можно найти по этой ссылке: [GitHub repository](https://github.com/yhat/currency-portfolio-optimization).

```python
yh = Yhat('', '', 'http://cloud.yhathq.com/')
yh.deploy('CurrencyPortfolio', CurrencyPortfolio, globals())
```

Перед тем как мы будем собирать инвестиционный портфель, давайте посмотрим на наши данные. Мы максимизируем возврат инвестиций в то время как минимизируем вероятность потерять деньги (дисперсия возврата инвестиций).

Если предположить, что рынок обмена валюты работает также как и любой другой рынок, тогда ожидаемый возврат инвестиций будет более обладать большей дисперсией по мере роста.

Построим график, чтобы убедиться в этом.

```python
%matplotlib inline
import seaborn
seaborn.regplot('mean', 'variance', exp_returns)
```

В левой части находится небольшой выброс. Если посмотреть в исходный набор данных, то это окажется Венесуэла, экономика которой  чувствует [себя не очень хорошо](http://www.reuters.com/article/2014/12/30/venezuela-economy-idUSL1N0UE1FY20141230). Как будет выглядеть этот график без Венесуэлы?

```python
exp_returns_without_ve = exp_returns[exp_returns.index != 'RXI_N.M.VE']
seaborn.regplot('mean', 'variance', exp_returns, ci=None)
seaborn.regplot('mean', 'variance', exp_returns_without_ve)
```

Таким образом, без этого выброса, большим значениям ожидаемой прибыли соотвествует большая дисперсия. Мы подтвердили нашу гипотезу и теперь можем продолжить собирать инвестиционный портфель.

## Использование модели

Вся суть прелести использования ScienceOps в том, что оно позволяет использовать модели как REST API. Но при попытке загрузки модели она не работала из-за технических работ, которые проводились на сервере, поэтому ниже мы просто решим задачу оптимизации на локальной машине.

Построим несколько инвестиционных портфелей с $\alpha$ от 0 до 20 с шагом по 0.5. 

```python
risk_aversion = [ra/2.0 for ra in range(41)]

def predict (data):
    P = matrix(data['risk_aversion'] * returns_cov.as_matrix())
    q = matrix(-exp_returns['mean'].as_matrix())
    G = matrix(0.0, (len(q),len(q)))
    G[::len(q)+1] = -1.0
    h = matrix(0.0, (len(q),1))
    A = matrix(1.0, (1,len(q)))
    b = matrix(1.0)

    solution = solvers.qp(P, q, G, h, A, b)
    expected_return = exp_returns['mean'].dot(solution['x'])[0]
    variance = sum(solution['x'] * returns_cov.as_matrix().dot(solution['x']))[0]

    investments = {}
    for i, amount in enumerate(solution['x']):
        # Ignore values that appear to have converged to 0.
        if amount > 10e-5:
            investments[countries[i]] = amount*100

    return {
        'risk_aversion': data['risk_aversion'],
        'investments': investments,
        'expected_return': expected_return,
        'variance': variance
    }
    

portfolios = []
for ra in risk_aversion:
    portfolios.append(predict({'risk_aversion': ra}))    
```

Это даёт нам представление о нашей [границе эффективности](http://en.wikipedia.org/wiki/Efficient_frontier) между прибылью и риском. В зависимости от желаемой степени риска мы можем выбирать различные варианты инвестирования.

```python
from matplotlib import pyplot
pyplot.plot(risk_aversion, [p['expected_return'] for p in portfolios], 'o-')
pyplot.axis([-1, 21, 0.0, 0.3])
pyplot.title('Efficient Frontier for Currency Portfolios')
pyplot.text(9, -0.05, 'Risk Aversion')
pyplot.text(-4, 0.18, '% Return', rotation='vertical')
```

Мы также можем оценить на баланс с точки зрения ожидаемой прибыли и дисперсии.

```python
pyplot.plot(
    [p['expected_return'] for p in portfolios],
    [p['variance'] for p in portfolios],
    'o-'
)
pyplot.axis([0, 0.3, -0.5, 7.5])
pyplot.title('Efficient Frontier for Currency Portfolios')
pyplot.text(.1, -2, 'Expected Return (%)')
pyplot.text(-0.025, 5, 'Portfolio Return Variance', rotation='vertical')
```

Теперь выведем все расссчитанные инвестиционный портфели. Эти портфели будут содержать информацию о том, какую долю денег вкладывать в различную валюту. К примеру, Портфель 4 подразумевает инвестирование 60.55%, 37.25% и 2.2% всей суммы в юани, гонконгские доллары и тайские баты, соотвественно.

```python
# We print out all the portfolios available to us.
from operator import itemgetter

for i, p in enumerate(portfolios):
    print 'Portfolio %d: expected monthly risk_aversion=%.1f, return=%.03f%%, variance=%.03f' % (
        i, p['risk_aversion'], p['expected_return'], p['variance']
    )
    for country, investment in sorted(p['investments'].items(), key=itemgetter(1), reverse=True):
        print '\t%5.02f%% %s' % (investment, country)
    print
    
```
